```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node("fact_rides")
```

```python
G.add_node("dim_driver")
G.add_node("dim_vehicle")
G.add_node("dim_customer")
G.add_node("dim_pickuptime")
```

```python
G.add_node("dim_customernum")
G.add_node("dim_vehicle_type")
G.add_node("dim_pickup_city")
G.add_node("dim_drop")
```

```python
G.add_edge("fact_rides","dim_driver")
G.add_edge("fact_rides","dim_vehicle")
G.add_edge("fact_rides","dim_customer")
G.add_edge("fact_rides","dim_pickuptime")
```

```python
G.add_edge("dim_driver","dim_pickup_city")
G.add_edge("dim_vehicle","dim_vehicle_type")
G.add_edge("dim_customer","dim_customernum")
G.add_edge("dim_pickuptime","dim_drop")
```

```python
pos = nx.spring_layout(G,seed=42)
plt.figure(figsize=(10,7))
nx.draw(G,pos,with_labels=True,node_size=3000,font_size=9)
plt.title("Snowflake Schema Representation")
plt.show()
```
